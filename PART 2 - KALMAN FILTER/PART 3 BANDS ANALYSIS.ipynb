```python
pnl_data_1 = pd.DataFrame()
pnl_data_2 = pd.DataFrame()
pnl_data_3 = pd.DataFrame()
pnl_data_4 = pd.DataFrame()
pnl_data_5 = pd.DataFrame()
pnl_data_6 = pd.DataFrame()
pnl_data_7 = pd.DataFrame()
pnl_data_8 = pd.DataFrame()
tickers = list(data.index)

def get_pnl(band, dbase, band_base):
    d = pd.read_csv(file_main_data + 'pair_sim_' + str(band) + '.csv', low_memory=False)
    dg = d.groupby('symbol')
    tickers = list(data.index)
    for i in range(len(dbase)):
        band_base[tickers[i]] = dg.get_group(tickers[i]).pnl.reset_index(drop=True)
    return band_base

def get_pnl_list(pnl_data, band):
    list_of_pnl = []
    for i in range(len(coint_pairs)):
        tick = pd.DataFrame()
        whole_avg = pnl_data[tickers[i]].mean()
        
        tick['data'] = pnl_data[tickers[i]]
        tick['pos'] = pnl_data[tickers[i]] > 0
        
        t = tick.groupby('pos')
        
        if len(t.get_group(False)) == len(tick):
            t_len = len(tick)
            tt_len = 0
                        
            t_mean = t.get_group(False)['data'].mean()
            tt_mean = 0
            
            r = (tickers[i], t_len, tt_len, t_mean, tt_mean, whole_avg, data['perf_'+str(band)].iloc[i])
            list_of_pnl.append(r)
            
        else:

            t_len  = len(t.get_group(False))
            tt_len = len(t.get_group(True))

            t_mean  = t.get_group(False)['data'].mean()
            
            tt_mean = t.get_group(True)['data'].mean()
            r = (tickers[i], t_len, tt_len, t_mean, tt_mean, whole_avg, data['perf_'+str(band)].iloc[i])
            list_of_pnl.append(r)


    return list_of_pnl        


def make_df_pnl(pnl_list, band):
    df = pd.DataFrame(data=pnl_list, columns=['pair', 'no_loss', 'no_win', 'avg_loss', 'avg_win', 'whole_avg', 'perf_'+str(band)])
    df.set_index('pair', inplace=True)
    
    s = list(df.index)
    
    df['stat'] = [s[i].split(':')[1] for i in range(len(s))]
    df['pos_avg'] = df.whole_avg > 0
    df['win'] = df['perf_'+ str(band)] > 100000
    tot_days = df['no_loss'] + df['no_win']
    df['loss_percent'] = round(df['no_loss']/tot_days, 2)
    df['win_percent'] = round(df['no_win']/tot_days, 2)
    
    
    return df[df['perf_'+str(band)] != 100000]


pnl_1 = get_pnl(0.25, data, pnl_data_1)
pnl_2 = get_pnl(0.5,  data, pnl_data_2)
pnl_3 = get_pnl(0.75, data, pnl_data_3)
pnl_4 = get_pnl(1,    data, pnl_data_4)
pnl_5 = get_pnl(1.25, data, pnl_data_5)
pnl_6 = get_pnl(1.5,  data, pnl_data_6)
pnl_7 = get_pnl(1.75, data, pnl_data_7)
pnl_8 = get_pnl(2,    data, pnl_data_8)


list_1 = get_pnl_list(pnl_1, 0.25)
list_2 = get_pnl_list(pnl_2, 0.5)
list_3 = get_pnl_list(pnl_3, 0.75)
list_4 = get_pnl_list(pnl_4, 1)
list_5 = get_pnl_list(pnl_5, 1.25)
list_6 = get_pnl_list(pnl_6, 1.5)
list_7 = get_pnl_list(pnl_7, 1.75)
list_8 = get_pnl_list(pnl_8, 2)


df_1 = make_df_pnl(list_1, 0.25)
df_2 = make_df_pnl(list_2, 0.5)
df_3 = make_df_pnl(list_3, 0.75)
df_4 = make_df_pnl(list_4, 1)
df_5 = make_df_pnl(list_5, 1.25)
df_6 = make_df_pnl(list_6, 1.5)
df_7 = make_df_pnl(list_7, 1.75)
df_8 = make_df_pnl(list_8, 2)



def get_tables(data, band):
    
    cols = ['no_loss', 'no_win', 'avg_loss','avg_win', 'whole_avg', 'perf_' + str(band), 'loss_percent', 'win_percent']
    
    avg_list = [list(round(data[cols].mean(), 2).reset_index(drop=True))[i] for i in range(8)]
    
    stat_list = list(round(data.groupby('stat').get_group('STAT')[cols].mean(), 2))
    
    nstat_list = list(round(data.groupby('stat').get_group('NON-STAT')[cols].mean(), 2))
    
    stat = data.groupby('stat').get_group('STAT')
    
    win_stat = list(round(stat.groupby('win').get_group(True)[cols].mean(), 2))
    
    nstat = data.groupby('stat').get_group('NON-STAT')
    
    win_nstat = list(round(nstat.groupby('win').get_group(True)[cols].mean(), 2))
    
    df_list = [avg_list, stat_list, nstat_list, win_stat, win_nstat]
    
    index = ['avg', 'avg stat', 'avg non-stat', 'avg winning stat', 'avg winning non-stat']
    
    list_to_df =  pd.DataFrame(data=df_list, columns=cols, index=index)
    list_to_df.to_latex(buf=file_data + str(band)+' daily returns band.txt', index=True)
    print(list_to_df)
    

get_tables(df_1, 0.25)
get_tables(df_2, 0.5)
get_tables(df_3, 0.75)
get_tables(df_4, 1)
get_tables(df_5, 1.25)
get_tables(df_6, 1.5)
get_tables(df_7, 1.75)
get_tables(df_8, 2)
```
